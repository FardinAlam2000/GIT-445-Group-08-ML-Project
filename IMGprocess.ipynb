```python
import os
from PIL import Image
from matplotlib import pyplot as plt
```

```python
root = 'images/'

fnames = os.listdir('images/')
```

```python
len(fnames)
```

```python
fig, axs = plt.subplots(nrows=10, ncols=5, figsize = (25, 20))

axs = axs.flatten()

for i in range(50):
  filepath = os.path.join(root, fnames[i])
  img = Image.open(filepath)
  axs[i].imshow(img)
  axs[i].axis('off')
  axs[i].set_title(fnames[i])
  
plt.show()
```
